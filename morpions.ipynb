```python
def afficher(L):
    print("      0)  1)  2)")
    print("    ------------")
    print("0)  |",L[0],"|",L[1],"|",L[2],"|")
    print("    ------------")
    print("1)  |",L[3],"|",L[4],"|",L[5],"|")
    print("    ------------")
    print("2)  |",L[6],"|",L[7],"|",L[8],"|")
    print("    ------------")
```

```python
def jouer(joueur,L):
    c = int(input("Entrez le numero de la colonne : "))
    l = int(input("Entrez le numero de la ligne : "))
    while(c>2 or c<0 or l>2 or l<0 or L[c+l*3]!=" ") :
        print("Coordonnee incorrecte ou Case NON VIDE ")
        print("Veuillez recommencer")
        c = int(input("Entrez le numero de la colonne : "))
        l = int(input("Entrez le numero de la ligne : "))
        
    print("Vous avez joué la case (",c,",",l,")")
    if (joueur == 1):
        L[c+l*3] = "X"
    if (joueur == 2):
        L[c+l*3] = "O"

    afficher(L)
```

```python
def gagné(L):

    # TEST  victoire avec ligne ok
    if ((L[0]!=" ")and (L[0]==L[1]) and (L[0]==L[2])):
        return 1
    if (L[3]!=" ") and (L[3]==L[4]) and (L[3]==L[5]) :
        return 1
    if (L[6]!=" ")and (L[6]==L[7]) and (L[6]==L[8]):
        return 1
    # TEST  victoire avec colonne ok
    if (L[0]!=" ") and (L[0]==L[3]) and (L[0]==L[6]):
        return 1
    if (L[1]!=" ") and (L[1]==L[4]) and (L[1]==L[7]):
        return 1
    if (L[2]!=" ") and (L[2]==L[5]) and (L[2]==L[8]):
        return 1
    # TEST  victoire avec diagonale ok
    if (L[0]!=" ") and (L[0]==L[4]) and (L[0]==L[8]):
        return 1
    if (L[2]!=" ") and (L[2]==L[4]) and (L[2]==L[6]):
        return 1
        
```

```python
def GrillePleine(L):
    """ Retourne 1 si la grille est pleine ,0 sinon
    """
    s = 0
    for i in range (0,9):
        if (L[i] != " "):
            s = s+1
    if (s == 9):
        return 1
    else :
        return 0
```
